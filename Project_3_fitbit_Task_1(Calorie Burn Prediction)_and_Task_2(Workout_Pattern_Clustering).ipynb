```python
Q1 = df['Calories_Burned (kcal)'].quantile(0.25)
Q3 = df['Calories_Burned (kcal)'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df['Calories_Burned (kcal)'] < lower) |
    (df['Calories_Burned (kcal)'] > upper)
]

print("Number of outliers:", len(outliers))
```

```python
sns.boxplot(x=df['Calories_Burned (kcal)'])
plt.show()
```

Intrepretation:

Calorie burned shows outliers, since it is target variable for ML  problem will do clipping using IQR method

```python
Q1 = df['Calories_Burned (kcal)'].quantile(0.25)
Q3 = df['Calories_Burned (kcal)'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['Calories_Burned (kcal)'] = df['Calories_Burned (kcal)'].clip(lower, upper)
```

Cleaning for USL-clustering

```python
cols_to_drop = [
    'Calories_Burned (kcal)',
    'Workout_Type',
    'Workout_Type_HIIT',
    'Workout_Type_Mixed',
    'Workout_Type_Strength',
    'Workout_Type_Yoga'
]

cluster_df = df.drop(
    columns=[c for c in cols_to_drop if c in df.columns]
)
```

```python
print(df.columns.tolist())
```

```python
cluster_df = df.drop(
    ['Calories_Burned (kcal)', 'Workout_Type'],# c-B is the target variable for regression and wt label does not use in clustering
    axis=1
)
```

Encoding

```python
cluster_df = pd.get_dummies(
    cluster_df,
    columns=['Gender'],
    drop_first=True
)
```

```python
print(cluster_df.head())
```

Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(cluster_df)
```

PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)
print("Explained Variance Ratio:")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)


print(pca.explained_variance_ratio_)
```

K-means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=55,
    n_init=10
)

clusters = kmeans.fit_predict(X_pca)
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_pca, clusters)

print("Silhouette Score:", score)
```

```python
cluster_df['Cluster'] = clusters
```

```python
cluster_df['Cluster'].value_counts()
```

```python
cluster_df.groupby('Cluster').mean()
```

Interpretation:

cluster 2 has workout_frequency and experience level as more rather than cluster 1 and 0 which results leads to low fat_percentage.

cluster 1 has low experience level as well workout frequency, and it clearly shows Fat_percentage as High compare to three cluster models
